# Training a model that predicts Air Quality Index Classifications

## Classifications are as such:<br>
* Good
* Moderate
* Unhealthy f. Sens. G.
* Unhealthy
* Very Unhealthy
* Hazardous

```python
# Import neccessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.impute import KNNImputer
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

import functions.uap as uap
```

```python
# Load the dataset
air_df = uap.get_data()
```

```python
air_df.head()
```

```python
air_df.describe()
```

```python
air_df_selected = uap.select_and_rename_columns(air_df, "target_aqi")
```

```python
# Preprocessing the data
## Statistics
air_df_selected.describe()
```

```python
## Assign the target and features variables
X = air_df_selected.copy().drop(columns=["target"], axis=1)
y = air_df_selected["target"]
y_con, y_labels = pd.factorize(air_df_selected["target"])
## Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_con, test_size=0.2,stratify=y, random_state=42)
```

```python
## Building a pipeline without SMOTE
pipeline = Pipeline([
    ('imputer', KNNImputer()), 
    ('scaler', MinMaxScaler()),
    ('clf', LogisticRegression()) #placeholder
])

param_grid = [
    {
        'clf': [LogisticRegression(max_iter=1000)],
        'clf__C': [0.1, 1, 10], #Regularisation strength
        'clf__solver': ['lbfgs'] #Optimizes algorithm
    },
    {
        'clf': [RandomForestClassifier()],
        'clf__n_estimators': [100, 200], #Number of trees
        'clf__max_depth': [10, 5] #Maximum depth of the tree
    },
    {
        'clf': [XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')],
        'clf__n_estimators': [100], #Number of trees
        'clf__max_depth': [3, 5],#Maximum depth of the tree
        'clf__learning_rate': [0.1] #Learning rate
    }
]
```

```python
# Perform a grid search
grid_search = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, scoring='f1_macro')
grid_search.fit(X_train, y_train)

# Show results
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```

```python
# Evaluate the model
uap.check_classification(grid_search, X_train, X_test, y_train, y_test, y_labels)
```

```python
## Building a pipeline with SMOTE, balanced class weights, and more hyperparameters tuning 
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
pipeline = Pipeline([
    ('imputer', KNNImputer()), 
    ('scaler', MinMaxScaler()),
    ('oversampler', SMOTE(sampling_strategy='not majority', random_state=42)),
    ('clf', LogisticRegression()) #placeholder
])

param_grid = [
    {
        'clf': [LogisticRegression(max_iter=1000, class_weight='balanced')],
        'clf__C': [0.1, 1, 10], #Regularisation strength
        'clf__solver': ['lbfgs'] #Optimizes algorithm
    },
    {
        'clf': [RandomForestClassifier(class_weight='balanced')],
        'clf__n_estimators': [100, 200], #Number of trees
        'clf__max_depth': [10, 5] #Maximum depth of the tree
    },
    {
        'clf': [XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', sample_weight='balanced')],
        'clf__n_estimators': [100, 200],
        'clf__max_depth': [3, 5, 7],
        'clf__learning_rate': [0.05, 0.1],
        'clf__subsample': [0.8, 1.0],
        'clf__colsample_bytree': [0.8, 1.0]
}
    
]
```

```python
# Perform a grid search
grid_search_smote = GridSearchCV(pipeline, param_grid, cv=5, n_jobs=-1, scoring='f1_macro')
grid_search_smote.fit(X_train, y_train)

# Show results
print("Best parameters:", grid_search_smote.best_params_)
print("Best cross-validation score:", grid_search_smote.best_score_)
```

```python
# Evaluate the model
uap.check_classification(grid_search_smote, X_train, X_test, y_train, y_test, y_labels)
```

```python
## Assign the target and features variables
X = air_df_selected.copy().drop(columns=["target"], axis=1)
y = air_df_selected["target"]
y_con, y_labels = pd.factorize(air_df_selected["target"])
## Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y_con, test_size=0.2,stratify=y, random_state=42)
```

```python
## Pipeline to reduce overfitting
# Build the pipeline
pipeline = Pipeline([
    ('imputer', KNNImputer()),
    ('scaler', MinMaxScaler()),
    ('oversampler', SMOTE(sampling_strategy='not majority', random_state=42)),
    ('clf', XGBClassifier(
        use_label_encoder=False, 
        eval_metric='mlogloss',    # or 'mlogloss' for multiclass
        random_state=42
    ))
])

# Set up parameter grid to control overfitting
param_grid = {
    'clf__max_depth': [3, 5, 7],            # Lower values reduce overfitting
    'clf__min_child_weight': [1, 5, 10],    # Higher values reduce overfitting
    'clf__gamma': [0, 0.5, 1],              # Higher values reduce overfitting
    'clf__subsample': [0.7, 0.8, 1.0],      # Lower values reduce overfitting
    'clf__colsample_bytree': [0.7, 0.8, 1.0], # Lower values reduce overfitting
    'clf__reg_alpha': [0, 0.1, 1],          # L1 regularization
    'clf__reg_lambda': [1, 1.5, 2],         # L2 regularization
    'clf__learning_rate': [0.01, 0.05, 0.1],# Lower values reduce overfitting
    'clf__n_estimators': [100, 200]         # More trees with lower learning rate
}

# Grid search with cross-validation
grid_search = GridSearchCV(
    pipeline, 
    param_grid, 
    cv=5, 
    scoring='f1_macro',    
    n_jobs=-1
)

# Fit the grid search to your data
grid_search.fit(X_train, y_train)

print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)
```
