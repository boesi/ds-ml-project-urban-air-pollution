# Explaratory Data Analysis

```python
# Importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as mno
```

```python
# Loading data
air_df = pd.read_csv("/Users/saidalkildani/Desktop/DS_AI/Projects/ds-ml-project/data/UAP_Africa.csv")
```

```python
air_df.describe()
```

## EDA - Formaldehyde

```python
# Create a datafram of formaldehyde features
air_form = air_df[['target','L3_HCHO_tropospheric_HCHO_column_number_density', 'L3_HCHO_tropospheric_HCHO_column_number_density_amf']]
```

```python
# Statistics
air_form.describe()
```

```python
# Missing values
air_form.isnull().sum()
```

```python
# Replace column names
air_form.columns = ['target', 'form_density', 'form_density_amf']
```

```python
# Do the missing values have a pattern?
mno.matrix(air_form, figsize=(25, 10))
## The missing values are missing from both columns. 
```

```python
air_form
```

```python
# Multicollinearity
# Scatter plotting 
sns.scatterplot(air_form, x='form_density', y='form_density_amf')
# Calculate the correlation
air_form[['form_density', "form_density_amf"]].corr()
```

```python
# Correlation with the target
sns.pairplot(air_form)
air_form.corr()['target']
```

### The two values do not seem to be correlated with each other and both are recommended as features
